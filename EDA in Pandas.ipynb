### Are there continents that have grown faster than the others

```python
df.groupby('Continent').mean(numeric_only = True).sort_values(by = '2022 Population', ascending = False)
```

```python
df[df['Continent'] == 'Oceania'].head()
```

```python
df2 = df.groupby('Continent')[['1970 Population', '1980 Population', '1990 Population',
       '2000 Population', '2010 Population', '2015 Population',
       '2020 Population', '2022 Population']].mean(numeric_only = True).sort_values(by = '2022 Population', ascending = False)
df2
```

```python
df3 = df2.transpose()
```

```python
import matplotlib.pyplot as plt

colors = {
    "Asia": "red",
    "Africa": "green",
    "Europe": "blue",
    "North America": "orange",
    "South America": "purple",
    "Oceania": "brown"
}

df3.plot(figsize=(20,10), color=[colors[c] for c in df3.columns])

plt.title("Population Growth by Continent (1970–2022)")
plt.xlabel("Year", fontsize = 16, fontweight = 'bold')
plt.ylabel("Population (average per country)",  fontsize = 16, fontweight = 'bold')
plt.legend(title="Continent", labels=df2.index, loc="best")

plt.show()
```
